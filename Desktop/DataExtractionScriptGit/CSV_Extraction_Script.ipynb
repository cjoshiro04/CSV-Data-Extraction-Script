```python
import re
import pandas as pd
import random

#generates a list of all tweets from a data set along with the tweets respective true label.
#use: tweet_list = tweet_list_generator("data.csv", "tweet", "hate_speech")
#note that it returns a list of lists where each sublist is [tweet, true_label]
def tweet_list_generator(filename, tweet_column, hate_speech_column):
    tweet_list = []#list of lists
    #sub_list = [] #each of these sublists stores [tweet, hs_score]

    #convert to dataframe
    df = pd.read_csv(filename)

    for i in range(len(df)):
            tweet = df.loc[i, tweet_column]
            score = df.loc[i, hate_speech_column]
            res = re.search('@\w+:', tweet)
            if (res != None): 
                sub_list = []
                tweet_extract = tweet.split(res.group())
                sub_list.append(tweet_extract[1])#add tweeet to sublist
                sub_list.append(score)#add hate speech score to sublist
                tweet_list.append(sub_list)
                
    return tweet_list


#returns a list of user objects of the user class
#randomly generates a number of tweets to assign to each user. 
#also, the tweet assigned to the user is random, not sequential.
def create_user_dict(tweet_list):
    random.seed()
    
    user_dict = {}
    j = 0

    while(len(tweet_list)!=0):
        temp_list = []
        #generate random number of tweets to assign a user
        numTweets = random.randint(12, 30)#this range can be changed, 10-30 

        for i in range(0, numTweets):
            if(len(tweet_list)==0):
                break
            tweetIndex = random.randint(0, (len(tweet_list)-1))
            temp_list.append(tweet_list[tweetIndex])
            tweet_list.pop(tweetIndex)

        #add key value pair to dict
        user_dict[str(j)] = temp_list
        
        j+=1
        
    return user_dict

def print_dict(dict_name):
    for i in range(0, len(dict_name)):
        name = str(i)
        tweets = dict_name[str(i)]
        print("user: " + name)
        print("\n")
        print ("tweets: ")
        print(tweets)
        print("\n")
        print("\n")
        print("\n")
```

```python
tweet_list = tweet_list_generator("data.csv", "tweet", "hate_speech")
user_dict = create_user_dict(tweet_list)

print_dict(user_dict)
```
